```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot a histogram
sns.histplot(df['Period'], kde=True)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.hist(df['Data_value'], bins=10, color='blue', alpha=0.7)
plt.title('Histogram of Data_value')
plt.xlabel('NumericColumn values')
plt.ylabel('Frequency')
plt.show()
```

```python
import seaborn as sns

sns.countplot(x='STATUS', data=df)
plt.title('Bar Chart of STATUS')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(df['STATUS'].value_counts(), labels=df['STATUS'].unique(), autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart of STATUS')
plt.show()
```

```python
plt.scatter(df['Data_value'], df['MAGNTUDE'], alpha=0.5)
plt.title('Scatter Plot of Data_value vs MAGNTUDE')
plt.xlabel('Data_value')
plt.ylabel('MAGNTUDE')
plt.show()
```

```python
sns.boxplot(x='STATUS', y='Period', data=df)
plt.title('Box Plot of Period by STATUS')
plt.show()
```

```python
plt.plot(df['Period'], df['MAGNTUDE'])
plt.title('Line Plot of MAGNTUDE over Period')
plt.xlabel('Period')
plt.ylabel('MAGNTUDE')
plt.show()
```
